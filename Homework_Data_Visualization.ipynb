# 1. Dari negara mana saja pengunjung hotel berasal? 

*   5 negara dengan jumlah booking selesai terbanyak
*   5 negara dengan jumlah booking selesai paling sedikit



```python
# 1. Filter Check-Out
data_checkout = data[data['reservation_status']=='Check-Out']
data_checkout
```

```python
# 2. Aggregation Total Visitor per Country
data_country_agg = data_checkout.groupby('country_origin').agg(count=pd.NamedAgg(column='country_origin',aggfunc='count'))
data_country_agg
```

```python
# 3. Sort Top 5 Total Visitor per Country
data_country_sorted = data_country_agg.sort_values(['count'],ascending=False)
data_country_sorted_low = data_country_agg.sort_values(['count'])
data_country_sorted
```

```python
# 4. Top 5 Data
data_top_5 = data_country_sorted.head(5).reset_index(0)
data_top_5_low = data_country_sorted_low.head(5).reset_index(0)
data_top_5_low
```

```python
# 5. Visualization Top 5 Total Visitor per Country
plt.figure(figsize=(14,9))
plt.bar(data_top_5['country_origin'], data_top_5['count'])
plt.title('Top 5 negara dengan jumlah booking selesai terbanyak', size=25)
plt.xlabel('Negara', size=15)
plt.ylabel('Jumlah Booking', size=15)
plt.show
```

```python
# 6. Visualization Lowest 5 Total Visitor per Country
plt.figure(figsize=(14,9))
plt.bar(data_top_5_low['country_origin'], data_top_5_low['count'])
plt.title('Top 5 negara dengan jumlah booking selesai paling sedikit', size=25)
plt.xlabel('Negara', size=15)
plt.ylabel('Jumlah Booking', size=15)
plt.show
```
